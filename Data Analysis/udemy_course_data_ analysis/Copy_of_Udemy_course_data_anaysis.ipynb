```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=30, kde=True, color='blue')
plt.xlabel("Price")
plt.ylabel("Number of Courses")
plt.title("Distribution of Course Prices")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='price', y='num_subscribers', data=df, alpha=0.6)
plt.xlabel("Price")
plt.ylabel("Subscribers")
plt.title("Subscribers vs Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='subject', data=df, palette='viridis')
plt.xlabel("Subjects")
plt.ylabel("Number of Courses")
plt.title("Courses per Subject")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.barplot(x="is_paid", y="num_subscribers", data=df, estimator=sum, palette="Set2")
plt.xlabel("Course Type (Free vs Paid)")
plt.ylabel("Total Subscribers")
plt.title("Subscribers by Course Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="level", y="num_subscribers", data=df, estimator=sum, palette="Set2")
plt.xlabel("level")
plt.ylabel(" no of Subscribers")
plt.title("Subscribers by Course Type")
plt.show()
```

```python
num_subscribers=df["num_subscribers"].max()
print(num_subscribers)
```

```python
#most popular course title
df[df['num_subscribers'].max()==df['num_subscribers']]['course_title']
```

```python
#display 10 most popular courses as per no of subscribers
top_10=df.sort_values(by="num_subscribers",ascending=False).head(10)
print(top_10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x="num_subscribers",
    y="course_title",
    data=top_10.sort_values("num_subscribers", ascending=False),
    palette="viridis"
)
plt.xlabel("Number of Subscribers")
plt.ylabel("Course Title")
plt.title("Top 10 Courses by Subscribers (Sorted)")
plt.show()
```

```python
reviews_by_subject = df.groupby("subject")["num_reviews"].mean().reset_index()
print(reviews_by_subject )
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="subject", y="num_reviews", data=reviews_by_subject, palette="viridis")
plt.xlabel("Subjects")
plt.ylabel("Total Number of Reviews")
plt.title("Total Reviews per Subject")
plt.xticks(rotation=45)
plt.show()
```

```python
#price effect the no of reviews
plt.figure(figsize=(8,5))
sns.scatterplot(x='price', y='num_reviews', data=df, alpha=0.6)
plt.xlabel("Price")
plt.ylabel("no of reviews")
plt.title(" Reviews vs Price")
plt.show()
```

```python
df['year']=df['published_timestamp'].dt.year
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="year", data=df, palette="Set2")
plt.xlabel("Year")
plt.ylabel("Number of Courses")
plt.title("Courses Distribution by Year")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('year')['subject'].value_counts()
```

```python
df['year'] = df['published_timestamp'].dt.year

course_growth = df.groupby('year').size()

plt.figure(figsize=(10,6))
plt.plot(course_growth.index, course_growth.values, marker='o', linestyle='-', color='blue')
plt.title("Growth of Udemy Courses Published Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Courses Published")
plt.grid(True)
plt.show()
```
